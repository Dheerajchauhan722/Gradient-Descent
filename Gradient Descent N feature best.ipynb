```python
def step_gradient(x_train,y_train,learning_rate,m):
    scope_m=np.zeros(len(m))
   
    
    n=len(x_train)
    
    for i in range(n):
        
        for j in range(len(m)):
            scope_m[j]+=(-2/n)*((y_train[i]-(x_train[i]*m).sum())*x_train[i][j])
        
       
    new_m=m-learning_rate*scope_m
    
    return new_m
```
